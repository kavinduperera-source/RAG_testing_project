```python
import numpy as np
import matplotlib.pyplot as plt

# Two rectangular pulses
def rect(t, T):
    return np.where(np.abs(t) <= T/2, 1, 0)

# Convolution manually
def convolve_rects(T1, T2, t):
    # Result is trapezoid/triangle
    T_min = min(T1, T2)
    T_max = max(T1, T2)
    
    # Triangular part
    tri = np.where(np.abs(t) < T_min, 
                   T_min - np.abs(t), 0)
    
    # Rectangular part (if different lengths)
    rect_part = np.where(np.abs(t) < (T_max - T_min)/2,
                         T_min, 0)
    
    return tri + rect_part

# Using numpy convolution
t = np.linspace(-2, 2, 1000)
x = rect(t, 1.0)  # Pulse width 1
h = rect(t, 0.5)  # Pulse width 0.5
y = np.convolve(x, h, mode='same') * (t[1]-t[0])

plt.figure(figsize=(10, 6))
plt.plot(t, x, label='x(t): Rectangular Pulse T=1.0')
plt.plot(t, h, label='h(t): Rectangular Pulse T=0.5')
plt.plot(t, y, label='y(t): Convolution Result', linewidth=2)
plt.title('Convolution of Two Rectangular Pulses')
plt.xlabel('Time t')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()
plt.show()





```
